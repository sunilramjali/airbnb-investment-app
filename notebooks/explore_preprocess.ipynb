```
%%sql -r dataframe_5
SELECT column_name, data_type
FROM airbnb_investment_db.information_schema.columns
WHERE table_schema = 'GOLD'
  AND table_name = 'DIM_PROPERTY_GROUP'
ORDER BY ordinal_position;
```

```
SELECT COUNT(DISTINCT property_group) AS distinct_property_groups
FROM airbnb_investment_db.gold.mart_property_group
WHERE UPPER(city) LIKE '%MANCHESTER%';
```

```
SELECT
  neighbourhood,
  COUNT(DISTINCT property_group) AS distinct_property_groups,
  LISTAGG(DISTINCT property_group, ', ') WITHIN GROUP (ORDER BY property_group) AS property_groups
FROM airbnb_investment_db.gold.mart_property_group
WHERE UPPER(city) LIKE '%MANCHESTER%'
GROUP BY neighbourhood
ORDER BY neighbourhood
```

```
%%sql -r dataframe_4
SELECT
  neighbourhood,
  property_group,
  listing_count,
  ROUND(100.0 * listing_count
        / SUM(listing_count) OVER (PARTITION BY neighbourhood), 1) AS pct_of_neighbourhood
FROM airbnb_investment_db.gold.mart_property_group
WHERE UPPER(city) LIKE '%MANCHESTER%'
ORDER BY neighbourhood, pct_of_neighbourhood DESC;
```
